```python
def gettext(key,ip):
    #headers={'User-Agent':random.choice(agentpool)}
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.91 Safari/537.36'}
    url="http://zhaopin.baidu.com/company?query="+key+"&&page=firm&subPage=jcxx"
    #ip=random.choice(ippool)
    proxies={'http':'http://'+ip}
    print(proxies)
    try:
        result = requests.get(url, headers= headers,proxies=proxies,timeout=5) #在请求中设定头
    except:
        print('无法获取网页')
        return 2,''
    
    soup=BeautifulSoup(result.text,'lxml')
    
    if soup.title.text[:2]=='\n_':
        return 3,''
    else:
        for i in range(len(soup.title.text)):
            if soup.title.text[i]=='_':
                t=soup.title.text[:i]
                break
        s1=soup.find('div',class_='top mb16')
        s2=soup.find('div',class_='left width-left')
        with open('E:/Py/com/data3/'+key+'.txt','w',encoding='utf8') as f:
            f.write(str(s1)+str(s2))
        return 1,t.strip()
```
