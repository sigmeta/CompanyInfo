```python
fc=open('E:/Py/com/cominfo1.txt','r',encoding='utf-8')
```

```python
cominfo=[]
for line in fc:
    ci=line.strip().split(',')
    cominfo.append(ci)
```

```python
cominfo1=cominfo
```

```python
comlist=[]
f=open('E:/Py/com/comlist.txt','r',encoding='utf-8')
for line in f:
    comlist.append(line[:-1])
f.close()
len(comlist)
```

```python
cominfo=[]
for ci in cominfo1:
    if ci[0] in comlist:
        cominfo.append(ci)
        
len(cominfo)
```

```python
for ci in cominfo:
    try:
        # 执行sql语句
        sql="INSERT INTO companyinfo VALUES ('%s','%s', '%s')"%(ci[0],ci[1],ci[2])
        #print("INSERT INTO companyinfo VALUES ('%s','%s', '%s')"%(ci[0],ci[1],ci[2]))
        cursor.execute(sql.encode('utf8'))
        # 提交到数据库执行
        db.commit()
    except:
        # 如果发生错误则回滚
        db.rollback()
        print(0,ci[0],ci[1],ci[2])
```

```python
sql = "INSERT INTO companyinfo VALUES ('地方', 'Mohan', 'M')"
try:
    # 执行sql语句
    cursor.execute(sql.encode('utf8'))
    # 提交到数据库执行
    db.commit()
except:
    # 如果发生错误则回滚
    db.rollback()
    print(0)
```
