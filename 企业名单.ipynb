```python
from sqlalchemy import create_engine
import pymysql
import pandas as pd
import re
```

```python
df=pd.read_csv('E:/Py/pyzp/jsonproc/new/rawdata.csv')
df
```

```python
dfc=df.loc[:,'officialname']
```

```python
dfc=dfc.drop_duplicates()
```

```python
dfc=dfc.reset_index(drop=True)
```

```python
dr=[]
for i in range(len(dfc)):
    try:
        st=dfc.loc[i]
        if re.search('某',st) or re.search('行业',st) or re.search('top',st) or re.search('XX',st) or re.search('TOP',st) or re.search('BAT',st) or re.search('#',st) or re.search('上市',st) or re.search('国内',st) or re.search('前\d',st) or re.search('\d强',st):
            dr.append(i)
        else:
            dfc.loc[i]=re.sub('\(','（',dfc.loc[i])
            dfc.loc[i]=re.sub('\)','）',dfc.loc[i])
    except:
        dr.append(i)
print(dr)
```

```python
dfc=dfc.drop(dr)
dfc=dfc.reset_index(drop=True)
```

```python
engine = create_engine('mysql+pymysql://root:9432@localhost:3306/pyzp?charset=utf8')
dfc.to_sql('companylist',engine,if_exists='replace')
```
