```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url='http://www.kuaidaili.com/free/intr/'
allip=[]
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.90 Safari/537.36"}

for i in range(1,3):
    r=requests.get(url+str(i)+'/',headers=headers)
    soup=BeautifulSoup(r.text,"lxml")
    #print(soup.prettify())
    
    for it in soup.find_all('td'):
        try:
            st=it.get_text()
            if len(st)>10 and len(st)<16 and st[0]>'0' and st[0]<'9':
                allip.append(st+':'+it.next_sibling.next_sibling.get_text())
                #print(st)
                #print(it.next_sibling.next_sibling.get_text())
        except:
            continue
print(len(allip),allip)
```
